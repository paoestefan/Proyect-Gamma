```python
#Cleaning & merge 2019,2020
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Study data files
path1 = "Resources/2019.csv"
path2 = "Resources/2020.csv"
# Read the mouse data and the study results
data1 = pd.read_csv(path1)

data2 = pd.read_csv(path2)

data1.head()
```

```python
data1["Country or region"].unique()
```

```python
data1=data1.rename(columns={"Country or region":"Country_name"})
```

```python
data1.head()
```

```python
data2.head()
```

```python
data2["Country name"].unique()
```

```python
data2=data2.drop(columns={"Standard error of ladder score","upperwhisker","lowerwhisker","Logged GDP per capita","Social support","Generosity","Perceptions of corruption","Ladder score in Dystopia","Dystopia + residual","Healthy life expectancy","Freedom to make life choices"})
data2.head()
```

```python
data2=data2.rename(columns={"Regional indicator":"Regional_indicator","Country name":"Country_name","Ladder score":"Score","Explained by: Log GDP per capita":"GDP per capita","Explained by: Social support":"Social support","Explained by: Healthy life expectancy":"Healthy life expectancy","Explained by: Freedom to make life choices":"Freedom to make life choices","Explained by: Generosity":"Generosity","Explained by: Perceptions of corruption":"Perceptions of corruption"})
data2.head()
```

```python
# Combine the data into a single dataset
merge_df = pd.merge(data2,data1, on="Country_name", how= "left",suffixes=('_2020', '_2019'))
merge_df.head()
```

```python
#finding the top 5 for 2020
top=merge_df["Score_2020"].nlargest()
top2_df=pd.DataFrame(top)
top2_df["Country_name"]=merge_df["Country_name"]
top2_df
#barplot
```

```python
plt.bar(top2_df["Country_name"],top2_df["Score_2020"], color={"red","blue","green","coral","lightblue"},alpha=0.5,  align="center")
plt.xticks(top2_df["Country_name"])
plt.title("Top 5 happy countries 2020")
plt.xlabel("Countries")
plt.ylabel("Score")

plt.show()
```

```python
#finding top 5 2019
top=merge_df["Score_2019"].nlargest()
top1_df=pd.DataFrame(top)
top1_df["Country_name"]=merge_df["Country_name"]
top1_df
#barplot
```

```python
plt.bar(top1_df["Country_name"],top1_df["Score_2019"], color={"red","blue","green","coral","lightblue"},alpha=0.5,  align="center")
plt.xticks(top1_df["Country_name"])
plt.title("Top 5 happy countries 2019")
plt.xlabel("Countries")
plt.ylabel("Score")

plt.show()
```

```python
tail=merge_df["Score_2019"].tail()
tail1_df=pd.DataFrame(tail)
tail1_df.sort_values(by='Score_2019',ascending=True)
tail1_df["Country_name"]=merge_df["Country_name"]
tail1_df
```

```python
plt.bar(tail1_df["Country_name"],tail1_df["Score_2019"], color={"red","blue","green","coral","lightblue"},alpha=0.5,  align="center")
plt.xticks(tail1_df["Country_name"])
plt.title("Last 5 happy countries 2019")
plt.xlabel("Countries")
plt.ylabel("Score")

plt.show()
```
