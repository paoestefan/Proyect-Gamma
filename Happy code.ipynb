```python
#Cleaning & merge 2019,2020
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Study data files
path1 = "Resources/2019.csv"
path2 = "Resources/2020.csv"
# Read the mouse data and the study results
data1 = pd.read_csv(path1)

data2 = pd.read_csv(path2)

data1.head()
```

```python
data1=data1.rename(columns={"Country or region":"Country_name"})
```

```python
data1.head()
```

```python
data2.head()
```

```python
data2=data2.drop(columns={"Standard error of ladder score","upperwhisker","lowerwhisker","Logged GDP per capita","Social support","Generosity","Perceptions of corruption","Ladder score in Dystopia","Dystopia + residual","Healthy life expectancy","Freedom to make life choices"})
data2.head()
```

## Analisis by region 

```python
#Correlation between factors & score per year
import scipy.stats as st
from scipy.stats import linregress
```

```python
Regional_df=merge_df.groupby(by=["Regional_indicator"]).mean()
Regional_df
```

## Correlation between factors and year score

```python
x_values=Regional_df["Score_2020"]
y_values=Regional_df["GDP per capita_2020"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
correlation_matrix = np.corrcoef(x_values, y_values)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
regress_values = x_values * slope + intercept
line_eq = f"y ={round(slope,2)} x  + {round(intercept,2)}"
plt.subplots(figsize=(15,10))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"red")
plt.annotate(line_eq,(6.0,0.6),fontsize=15,color="red")
plt.ylabel("GDP")
plt.xlabel("Score")
plt.title(f"The r squared for 2020 is :{r_squared,4}")
plt.savefig("Images/Score 2020 vs GDP.png")
plt.grid()
plt.show()
```

#### There is a strong positive correlation between GDP per capita and Happiness Score for year 2020

```python
x_values=Regional_df["Score_2019"]
y_values=Regional_df["GDP per capita_2019"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
correlation_matrix = np.corrcoef(x_values, y_values)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
regress_values = x_values * slope + intercept
line_eq = f"y ={round(slope,2)} x  + {round(intercept,2)}"
plt.subplots(figsize=(15,10))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"red")
plt.annotate(line_eq,(6.0,0.6),fontsize=15,color="red")
plt.ylabel("GDP")
plt.xlabel("Score")
plt.title(f"The r squared for 2019 is :{r_squared,4}")
plt.savefig("Images/Score 2019 vs GDP.png")
plt.grid()
plt.show()
```

#### There is a strong positive correlation between GDP per capita and Happiness Score for year 2019
